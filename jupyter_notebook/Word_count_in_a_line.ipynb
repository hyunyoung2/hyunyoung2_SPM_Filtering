```python
testPath=["/home/hyunyoung2/My_lab/smart_conference/data/Text/test/test_HAM-asp-utf8-5000aa",
        "/home/hyunyoung2/My_lab/smart_conference/data/Text/test/test_SPM-asp-utf8-5000aa"]
trainPath=["/home/hyunyoung2/My_lab/smart_conference/data/Text/train/train_20170616-asp-utf8-5000.ham_aa",
        "/home/hyunyoung2/My_lab/smart_conference/data/Text/train/train_20170616-asp-utf8-5000.spm_aa"]


def readAFileWithoutNewline(file):
    f = open(file, "r")
    
    # Because of iconv program, Error a little happens in converting EUC-KR(or CP949) to UTF-8
    lineByLine = [x for x in f.readlines() if x != "\n"]    
    
    # Remove Unicode space 
    for lineIdx, lineStr in enumerate(lineByLine):
        lineByLine[lineIdx] = tuple(lineStr.split())

    f.close()
    
    return lineByLine

def main():

    numOfWordsWithoutDuplication = 0
    numOfWords = 0
    numOfLines = 0
    maxOfwords = -1
    minOfwords = 99999
    for fileIdx, fileVal in enumerate(testPath):
        sentences = readAFileWithoutNewline(fileVal)
        numOfLines += len(sentences)
        numOfWordsWithoutDuplicationFile = 0
        numOfWordsFile = 0
        print("sentence:", sentences[0:2])
        for senIdx, senVal in enumerate(sentences):
            if maxOfwords < len(senVal):
                maxOfwords = len(senVal)
            if minOfwords > len(senVal):
                minOfwords = len(senVal)
            if len(senVal) == 243:
                print("test:", senVal)
            if senIdx <= 2:
                print(type(senVal), senVal)
            numOfWordsWithoutDuplicationFile += len(list(set(senVal)))
            numOfWordsFile += len(senVal)
            numOfWordsWithoutDuplication += len(list(set(senVal)))
            numOfWords += len(senVal)
        print("doc:", fileVal)
        print("without duplication:", numOfWordsWithoutDuplicationFile)
        print("with duplication:", numOfWordsFile)
        
            
    for fileIdx, fileVal in enumerate(trainPath):
        sentences = readAFileWithoutNewline(fileVal)
        numOfLines += len(sentences)
        numOfWordsWithoutDuplicationFile = 0
        numOfWordsFile = 0
        print("sentence:", sentences[0:2])
        for senIdx, senVal in enumerate(sentences):
            if maxOfwords < len(senVal):
                maxOfwords = len(senVal)
            if minOfwords > len(senVal):
                minOfwords = len(senVal)
            if len(senVal) == 243:
                print("test:", senVal)
            if senIdx <= 2:
                print(type(senVal), senVal)
            numOfWordsWithoutDuplicationFile += len(list(set(senVal)))
            numOfWordsFile += len(senVal)
            numOfWordsWithoutDuplication += len(list(set(senVal)))
            numOfWords += len(senVal)
        print("doc:", fileVal)
        print("without duplication:", numOfWordsWithoutDuplicationFile)
        print("with duplication:", numOfWordsFile)
            
    print("\nthe total of word:", numOfWords)
    print("\nthe total of word without duplication:", numOfWordsWithoutDuplication)
    
    print("\nthe total of lines:", numOfLines)
    
    print("\nminOfwords :", minOfwords)
    print("\nmaxOfwords :", maxOfwords)
    
if __name__ == "__main__":
    main()
        
        

        
```
